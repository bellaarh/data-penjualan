```python
import pandas as pd

df = pd.read_csv('data_bersih.csv')

print("=== DATA BERHASIL DIMUAT ===")
```

```python
total_omset = df['Paid Fees'].sum()
print(f"Total Pendapatan: USD {total_omset:,.0f}")

status_transaksi = df['Fees Status'].value_counts(normalize=True) * 100
print("\nPersentase Status Transaksi:")
print(status_transaksi)
```

```python
omset_per_channel = df.groupby('Advertising Channel')['Paid Fees'].sum().sort_values(ascending=False)
print("\nPendapatan per Saluran Iklan:")
print(omset_per_channel)

transaksi_per_channel = df.groupby('Advertising Channel')['Fees Status'].value_counts()
print("\nStatus Transaksi per Saluran Iklan:")
print(transaksi_per_channel)
```

```python
sales_performance = df.groupby('Sale Team')['Paid Fees'].sum().sort_values(ascending=False)
print("\nPerforma Sales Team (Total Pendapatan):")
print(sales_performance)

consultant_performance = df.groupby('Consultant')['Paid Fees'].sum().head(5)
print("\nTop 5 Konsultan berdasarkan Pendapatan:")
print(consultant_performance)
```

```python
telepon_vs_status = df.groupby('Fees Status')['Number of phone calls'].mean()
print("\nRata-Rata Jumlah Telepon per Status Pembayaran:")
print(telepon_vs_status)
```
